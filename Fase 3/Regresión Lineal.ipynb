```python
#importanto bibliotecas
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("datos mas usables.csv")
```

```python
data
```

```python
#Transformando a Dataframe
df = pd.DataFrame(data)
```

```python
#escogiendo variable dependiente e independiente

x = df[["Selling_Price"]]
y = df["Present_Price"]

#creando modelo regresin lineal

modelo = LinearRegression()

#ajustar el modelo a los datos

modelo.fit(x, y)

#generando coeficientes
beta_0 = modelo.intercept_
beta_1 = modelo.coef_

#Mostrar los coeficientes

print(f"intercepto(): {beta_0}")
print(f"coeficiente(): {beta_1}")

```

```python
plt.scatter(df["Selling_Price"], y, color = "blue", label = "datos reales")
plt.plot(df["Selling_Price"],  modelo.predict(x), color="red", label = "Linea de Regresion")
plt.title("Regresion Lineal: Kilometraje vs Precio de Venta")
plt.xlabel("Precio de Venta")
plt.ylabel("Precio actual")
plt.legend()
plt.show()
```

```python
#predecir precios para nuevos Valores
nuevo_x = np.array([[55], [65]])
prediccion = modelo.predict(nuevo_x)

#Mostrar las predicciones
print("\nPredicciones: ")
for i, pred in enumerate(prediccion):
    print(f"predicciÃ³n {i+1} para (Precio Actual: {nuevo_x[i][0]}$): {pred:.2f}$")

```

```python
#Variables independientes y dependientes
x = df [["Selling_Price", "Owner"]]
y = df ["Present_Price"]

#Crear un modelo de regresion lineal

modelo = LinearRegression()

modelo.fit(x, y)

#coeficientes regresion
beta_0 = modelo.intercept_
beta_1, beta_2 = modelo.coef_

#mostrar coeficientes

print(f"intercepto:(B0) {beta_0}")
print(f"coeficiente:(precio inicial) {beta_1}")
print(f"coeficiente:(propietarios) {beta_2}")


#visualizar la regresion

plt.scatter(df["Selling_Price"], y, color = "blue", label = "datos reales")
plt.plot(df["Selling_Price"],  modelo.predict(x), color="red", label = "Linea de Regresion")
plt.title("Regresion Lineal: Kilometraje vs Precio de Venta")
plt.xlabel("Precio de Venta")
plt.ylabel("Precio actual")
plt.legend()
plt.show()
```

```python
#predecir precios para nuevos Valores
nuevo_x = np.array([[55, 2], [65, 1]])
prediccion = modelo.predict(nuevo_x)

#Mostrar las predicciones
print("\nPredicciones: ")
for i, pred in enumerate(prediccion):
    print(f"predicciÃ³n {i+1} para (Precio Actual: {nuevo_x[i][0]}$), Propietarios: {nuevo_x[i][1]}): {pred:.2f}$")
```

```python
#importar librerias para ver rango de error
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

#datos
x = df [["Selling_Price", "Owner"]]
y = df ["Present_Price"]

#Crear un modelo de regresion lineal

modelo = LinearRegression()

modelo.fit(x, y)

#realizar predicio
prediccion = modelo.predict(x)

#metricas de evaluaciÃ³n
r2 = r2_score(y, prediccion)
mae =mean_absolute_error(y, prediccion)
mse= mean_squared_error(y,prediccion)
rnse = np.sqrt(mse)

#mostrar metricas
print(f"R 2(coeficiente de determincaciÃ³n): {r2}")
print(f"Error absoluto medio: {mae}")

```
